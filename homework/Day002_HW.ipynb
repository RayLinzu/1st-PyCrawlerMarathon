```python
import csv

data = []
# 開啟 CSV 檔案
with open('D:\web crawler\example.csv', newline='',encoding="utf-8") as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for i in rows:
        print(i[5])
        data.append(i)
```
