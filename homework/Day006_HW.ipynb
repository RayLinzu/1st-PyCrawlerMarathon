```python
import requests
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'}
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers',headers=headers)
response = r.text

response
```

### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間
import pandas as pd
import json
data = json.loads(response)
df = pd.DataFrame(data['data'])
df
```

```python
df_time = pd.DataFrame(df['question'].values.tolist())
df_time
```

```python
from datetime import datetime #轉換成時間

created_time = []
for time in df_time['created']:
    created_time.append(datetime.fromtimestamp(time))
df_time.insert(8, 'created_time', created_time)
df_time
```

```python
#發問時間
df_time['created_time']
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間

print('第一筆回答的時間{}'.format(df_time['created_time'][0]))
print('最後一筆回答的時間{}'.format(df_time['created_time'][4]))
```
