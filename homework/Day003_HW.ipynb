## xtmltodict 解析檔案內容

```python
import xmltodict

with open("D:\\web crawler\\data\\10002_Week24_CH.xml", "r", encoding='utf-8') as fd:
    doc = dict(xmltodict.parse(fd.read()))
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
with open("D:\\web crawler\\data\\64_72hr_CH.xml", "r", encoding="utf-8") as fh:
    doc = dict(xmltodict.parse(fh.read()))
```

```python
print(doc['cwbopendata']['dataset']['locations']['location'])
```

```python
target = doc['cwbopendata']['dataset']['locations']['location']
```

```python
# 1. 請問高雄市有多少地區有溫度資料？
print(len(target))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for value in target:
    print(value['locationName'])
    print(value['weatherElement'][0]['time'][0]['dataTime'] )
    print(value['weatherElement'][0]['time'][0]['elementValue'] )
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
target_first = target[0]
print(target_first['locationName'])
for value in target_first['weatherElement'][0]['time']:
    print(value['dataTime'])
    print(value['elementValue'])
```
