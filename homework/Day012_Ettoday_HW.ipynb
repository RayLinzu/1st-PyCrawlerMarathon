### ① 取出今天所有的發文

```python
import requests
import datetime
import re
from bs4 import BeautifulSoup

pattern = '[0-9]{4}\/[0-9]{2}\/[0-9]{2}'
today = datetime.date.today().strftime("%Y/%m/%d")

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

result = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    result.append([d.find(class_="date").text, d.find("em").text, d.find_all('a')[-1].text])
    #print(d.find(class_="date").text)
    if re.search(pattern, d.find(class_="date").text)[0] == today:
        print(d.find(class_="date").text, d.find("em").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
columns = ['時間', '類別', '標題']
target = pd.DataFrame(result, columns=columns)
target
```

```python
target.sort_values('類別')
```

### ③ 哪一個類別的文章最多

```python
df = pd.DataFrame(target.groupby('類別').size(), columns=['類別數量'])
df.sort_values('類別數量', ascending=False)
```
